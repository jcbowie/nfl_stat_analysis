```python
import requests
from pprint import pprint as pp
```

```python
# "http://www.nfl.com/feeds-rs/scores/2019/REG/6.json"
# /weeklyGamesAggrStats
url = "http://www.nfl.com/feeds-rs/weeklyGamesAggrStats/2018/REG/5.json"
url2 = "http://www.nfl.com/feeds-rs/teams.json"
url3 = "http://www.nfl.com/feeds-rs/schedules/.json"
url4 = "http://www.nfl.com/feeds-rs/teamQuickStats/0200.json"
url5 = "http://www.nfl.com/feeds-rs/teamSeasonStats/availableSeasons/ATL.json"
url6 = "http://www.nfl.com/feeds-rs/schedules/2018.json"
```

```python
response = requests.get(url6)
print(response)
```

```python
results = response.json()
```

```python
pp(results)
```

```python
games = results['gameSchedules']
```

```python
len(games)
```

```python
stats = []
url_game_preface = "http://www.nfl.com/feeds-rs/scores/byGame/"
for game in games:
    intrim_list = []
    for item in game.items():
        if (item[0] == "visitorTeam") | (item[0] == "homeTeam"):
            continue
        elif (item[0] == "gameId"):
            game_id = item[-1]
            game_url = "{}{}.json".format(url_game_preface, game_id)
            game_resp = requests.get(game_url)
            game_results = game_resp.json()
            pp(game_results)
            print()
            intrim_list.append(item[-1])
        elif (item[0] == "site"):
            for it in item[-1].items():
                intrim_list.append(it[-1])
        else:
            intrim_list.append(item[-1])
    stats.append(intrim_list)
```
