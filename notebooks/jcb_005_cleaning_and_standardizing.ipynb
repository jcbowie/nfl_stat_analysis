# Importing and cleaning
Action taken:
 - Importing the full game data and selecting only the records that are regular season games. <br>
 - Drop all of the columns that have NaN values. After review they arent neccesary for this process. <br>
 - Reset index

```python
df = pd.read_csv("../data/interim/game_data_full.csv")
df_reg = df.loc[df['seasonType'] == "REG"]
```

```python
print(len(df), "records in the original dataframe.")
print(len(df_reg), "records in the regular season games dataframe.")
print(len(df_reg.columns), "columns in the reg df.")
```

```python
for col, col_data in df_reg.iteritems():
    nan_count = col_data.isna().sum()
    if nan_count > 0:
        if col == "siteState":
            continue
        else:
            df_reg = df_reg.drop(col, axis=1)
```

```python
print(len(df_reg.columns), "columns in the reg df.")
```

```python
df_reg = df_reg.reset_index()
df_reg = df_reg.drop('index', axis=1)
```

```python
columns_not_standardized = ['gameDate', 'gameTimeEastern', 'homeNickname', 'visitorNickname',
                            'siteCity', 'siteState', 'roofType']
```

```python
df_objects = df_clean[columns_not_standardized]
df_clean = df_clean.drop(columns_not_standardized, axis=1)
```

```python
df_objects.head()
```

```python
def get_outcomes(df_row):
    if df_row['visitorTeamScore'] == df_row['homeTeamScore']:
        value = "D"
    elif df_row['visitorTeamScore'] > df_row['homeTeamScore']:
        value = "V"
    else:
        value = "H"
    return value
```

```python
df_clean['outcomes'] = df_clean.apply(get_outcomes, axis=1)
```

```python
df_clean.head()
```

```python
df_outcomes = df_clean['outcomes']
df_clean = df_clean.drop('outcomes', axis=1)
```
